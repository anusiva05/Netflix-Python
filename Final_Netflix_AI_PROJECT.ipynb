# Why Data Cleaning?
*  Cleaning removes errors → gives accurate insights

## Handling Null values

```python
# Missing values fill
df.fillna("Unknown", inplace=True)

# 1. Director → "Unknown"
df['director'] = df['director'].fillna("Unknown")

# 2. Cast → "unknown"
df['cast'] = df['cast'].fillna("unknown")

# 3. Country → "unknown"
df['country'] = df['country'].fillna("unknown")

# 4. Date Added → Convert + Mode
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Fill missing with most frequent date
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])

# 5. Rating → Mode
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])

# 6. Duration → "unknown"
df['duration'] = df['duration'].fillna("unknown")
```

# **5.Finding Outliers**

# Why Outlier


*   Outliers are extreme or unusual values in the dataset
*   They can distort analysis results (like average and trends)
*  Handling them helps improve accuracy and reliability of insights



```python
 # Select numeric columns
num_cols = ['release_year']

# Loop through columns and plot boxplot
for col in num_cols:
    fig = px.box(df, y=col, title=f"{col} Outliers")
    fig.show()
```

## Fix Outliers

```python
# Calculate quartiles
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)

# IQR formula
IQR = Q3 - Q1

# Filter data
df = df[(df['release_year'] >= Q1 - 1.5 * IQR) &
        (df['release_year'] <= Q3 + 1.5 * IQR)]
```

# Outliers in Box Plot

```python
# Box plot for release year
fig = px.box(df, y='release_year',
             title="Box Plot - Release Year")

fig.show()
```

## Z-Score

# Why Z-Score


*   It helps identify outliers by measuring how far a value is from the mean
*   It standardizes data, making values comparable across different scales
*  It improves data analysis and model accuracy by detecting unusual values




```python
# Calculate Z-score

# Convert to numeric
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

# Calculate Z-score
df['z_score'] = np.abs(stats.zscore(df['release_year'].dropna()))

# View result
df[['release_year', 'z_score']].head()
```

# **12.Automation**

# why Automation


*   Saves Time – Repetitive tasks are done quickly without manual effort.
*  Reduces Errors – Minimizes human mistakes and improves accuracy.
*  Increases Efficiency – Handles large data and processes faster and more consistently.



```python
import pandas as pd
import numpy as np
from scipy import stats
import plotly.express as px

def auto_data_analysis(df):

    print("🔹 Initial Shape:", df.shape)

    # -------------------------------
    # 1. DATA CLEANING
    # -------------------------------
    print("\n Data Cleaning")
    df.columns = df.columns.str.strip()

    # -------------------------------
    # 2. REMOVE DUPLICATES
    # -------------------------------
    print("\n Removing Duplicates")
    df = df.drop_duplicates()
    print("After removing duplicates:", df.shape)

    # -------------------------------
    # 3. NULL VALUES + HANDLING
    # -------------------------------
    print("\n Handling Missing Values")

    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna("Unknown")
        else:
            df[col] = df[col].fillna(df[col].median())

    print(df.isnull().sum())

    # -------------------------------
    # 4. DATA TYPE CONVERSION
    # -------------------------------
    print("\n Data Type Conversion")

    for col in df.columns:
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            pass

    print(df.dtypes)

    # -------------------------------
    # 5. OUTLIER DETECTION (Z-SCORE)
    # -------------------------------
    print("\n Outlier Detection")

    num_cols = df.select_dtypes(include=np.number).columns

    for col in num_cols:
        z = np.abs(stats.zscore(df[col], nan_policy='omit'))
        df = df[z < 3]

    print("After removing outliers:", df.shape)

    # -------------------------------
    # 6. TREND ANALYSIS
    # -------------------------------
    print("\n Trend Analysis")

    for col in num_cols:
        try:
            trend = df[col].value_counts().sort_index()
            px.line(trend, title=f"Trend - {col}").show()
        except:
            pass

    # -------------------------------
    # 7. HYPOTHESIS TESTING
    # -------------------------------
    print("\n Hypothesis Testing")

    if len(num_cols) >= 2:
        col1 = num_cols[0]
        col2 = num_cols[1]

        t, p = stats.ttest_ind(df[col1], df[col2])

        print(f"T-test between {col1} & {col2}")
        print("P-value:", p)

    else:
        print("Not enough numeric columns for hypothesis testing")

    print("\n Process Completed")

    return df


# -------------------------------
#  RUN FOR ANY DATASET
# -------------------------------
df = pd.read_csv("netflix_titles.csv")

df_clean = auto_data_analysis(df)
```
